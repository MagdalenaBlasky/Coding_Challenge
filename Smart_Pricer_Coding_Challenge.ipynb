# Smart Pricer Coding Challenge

1. Give an overview of the unique values for every column including their corresponding revenue and the amount of tickets sold. Store the results as an individual csv file for each column in a folder named “output”.

2. Give an overview of the revenue generated per date. Sorted in ascending order. Store the results as a csv file named “fingerprint.csv” in a folder named “output”.

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/data_export.csv',encoding='cp1252',sep=';')
data
```

```python
data.info()
```

```python
data.drop(columns=['YRES_ID'], inplace=True)
```

```python
for column in data.columns:
    print(column,':',data[column].unique())
```

```python
data['price'] = data['price'].str.replace(',','.')
data['price'] = data['price'].astype('float')
```

```python
data
```

## Task 1

```python
columns_for_calculation = ['ticket_types','person_types','event_date','channel','preiscode']

for i in columns_for_calculation:
    print(i,':',data[i].unique())
```

```python
def group_sum_and_count(column_name):
    
    group = data.groupby(column_name)

    df1 = group['price'].sum().reset_index(name='revenue')
    df2 = group[column_name].count().reset_index(name='count')
    df1['amount_sold'] = df2['count']
    
    fullpath = 'output/'+ column_name +'.csv'
    
    return df1.to_csv(fullpath,index=False)
```

```python
map(group_sum_and_count,columns_for_calculation)
```

## Task 2

```python
data.purchase_dt = pd.to_datetime(data.purchase_dt)
```

```python
data['purchase_date'] = data['purchase_dt'].dt.date
data['purchase_time'] = data['purchase_dt'].dt.time

data.drop(columns=['purchase_dt'], inplace=True)
```

```python
data.purchase_date.nunique()
```

```python
fingerprint = data.groupby('purchase_date').sum()
fingerprint.reset_index(inplace=True)
```

```python
fingerprint
```

```python
fingerprint.rename(columns={'price':'revenue'},inplace=True)
```

```python
fingerprint.to_csv('output/fingerprint.csv', index=False)
```
